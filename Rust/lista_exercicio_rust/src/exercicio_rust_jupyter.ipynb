```rust
use std::io;
```

```rust
//1. Olá Terráqueos
fn exercicio1() {
    println!("Olá terráqueos!!");
}
exercicio1();
```

```rust
//2. Montar número
fn exercicio2(numero: i32) {
    println!("O número informado foi: {}", numero);
}    
exercicio2(42);
```

```rust
//3. Soma de dois numeros
fn exercicio3(num1: i32, num2: i32) {
    let soma = num1 + num2;
    println!("A soma de {} e {} é: {}", num1, num2, soma);
}
exercicio3(9, 13);
```

```rust
//4. Média de 4 números
fn exercicio4(num1: f64, num2: f64, num3: f64, num4: f64) {
    let media = (num1 + num2 + num3 + num4) / 4.0;
    println!("A média de {}, {}, {} e {} é: {}", num1, num2, num3, num4, media);
}
exercicio4(10.0, 20.0, 30.0, 42.4);
```

```rust
//5. Metros para centímetros
fn exercicio5(metros: f64) {
    let centimetros = metros * 100.0;
    println!("{} metros equivalem a {} centímetros", metros, centimetros);
}
exercicio5(1.3);
```

```rust
//6. Área do Circulo    
fn exercicio6(raio: f64) {
    let area = std::f64::consts::PI * raio.powi(2);
    println!("A área do círculo com raio {} é: {}", raio, area);
}
exercicio6(3.0);
```

```rust
//7. Área do Quadrado e o dobro
fn exercicio7(lado: f64) {
    let area = lado.powi(2);
    let dobro = area * 2.0;
    println!("A área do quadrado com lado {} é: {} e o dobro é: {}", lado, area, dobro);
}   
exercicio7(2.0);
```

```rust
//8. Salário Mensal
fn exercicio8(salario_hora: f64, horas_trabalhadas: f64) {
    let salario_mensal = salario_hora * horas_trabalhadas;
    println!("O salário mensal é: {}", salario_mensal);
}
exercicio8(48.0, 200.0);
```

```rust
//9. Fahrenheit para Celsius
fn exercicio9(fahrenheit: f64) {    
    let celsius = (fahrenheit - 32.0) * 5.0 / 9.0;
    println!("{} graus Fahrenheit equivalem a {} graus Celsius", fahrenheit, celsius);
}
exercicio9(50.0);
```

```rust
//10. Celcius para Fahrenheit
fn exercicio10(celsius: f64) {
    let fahrenheit = (celsius * 9.0 / 5.0) + 32.0;
    println!("{} graus Celsius equivalem a {} graus Fahrenheit", celsius, fahrenheit);
}
exercicio10(40.0);
```

```rust
//11. Operações Matemáticas Mistas
fn exercicio11(n1: i32, n2: i32, n3: f64) {
    let r1 = (2 * n1) as f64 * (n2 as f64 / 2.0);
    let r2 = (3 * n1) as f64 + n3;
    let r3 = n3.powi(3);
    println!("A: O produto do dobro do primeiro com metade do segundo é: {}", r1);
    println!("B: A soma do triplo do primeiro com o terceiro é: {}", r2);
    println!("C: O terceiro elevado ao cubo é: {}", r3);    
}
exercicio11(10, 20, 5.0);
```

```rust
//12. Peso Ideal
fn exercicio12(altura: f64) {
    let peso_ideal = (72.7 * altura) - 58.0;
    println!("O peso ideal para a altura {} é: {}", altura, peso_ideal);
}
exercicio12(1.81);
```
