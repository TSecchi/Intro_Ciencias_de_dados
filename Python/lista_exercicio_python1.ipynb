```python
# Pedir 4 notas e mostrar a média

nota1 = float(input("escreva a Nota 1"))
nota2 = float(input("escreva a Nota 2"))
nota3 = float(input("escreva a Nota 3"))
nota4 = float(input("escreva a Nota 4"))

round = (nota1 + nota2 + nota3 + nota4) / 4
print ("A média da nota é:", round)
```

```python
# Converter metros para centímetros

metros = float(input("Digite o valor em metros: "))
centimetros = metros * 100

print ("Valor em centímetros =", centimetros)
```

```python
# Calcular área de um círculo

import math 
raio = float(input("Digite o raio do círculo"))
area = math.pi * raio**2
print ("A área do círculo é =", area)
```

```python
# Área do quadrado e o dobro da área

lado = float (input("Digite o lado do quadrado:"))
area = lado **2
dobro = area *2
print ("Área dp quadrado =", area)
print ("Dobro da área = ", dobro)
```

```python
# Operação com 2 numeros inteiros e um numero real

x1 = int(input("Digite o número inteiro: "))
x2 = int(input("Digite o número inteiro: "))
x3 = float(input("Digite um número real: "))

resultado1 = (2 * x1) * (x2 / 2)
resultado2 = (3 * x1) + x3
resultado3 = x3 ** 3

print ("Produto do dobro do primeiro com metade do segundo:", resultado1)
print ("Soma do triplo do primeiro com o terceiro",resultado2)
print ("Terceiro elevado ao cubo", resultado3)
```
